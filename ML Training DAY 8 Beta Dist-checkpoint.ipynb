```python
import numpy as np
import scipy.stats as stats

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import beta
data_beta = beta.rvs (a=0.1, b=0.1, size=10000)
```

```python
ax = sns.distplot(data_beta, kde=True, bins = 100, color='maroon', hist_kws = {"linewidth": 15, "alpha": 1})
ax.set (xlabel = "Beta Distribution", ylabel = "Probability")
plt.plot()
```
