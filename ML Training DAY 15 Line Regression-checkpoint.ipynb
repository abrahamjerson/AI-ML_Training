```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use(['ggplot'])
```

```python
#Create data

X = 2*np.random.rand(100,1)
y = 4+3 * X+np.random.randn(100,1)
plt.plot(X, y, 'b.')
plt.xlabel("$X$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
_=plt.axis([0,2,0,15])
```

```python
"""
fit_intercept: Whether to calculate the intercept for this model. If set to False, no intercept will be used in calculations

fit_intercept: This parameter is ignored when fit_intercept is set to False. If True, the regressors X will be normalized before regression by subtracting the mean and dividing by the l2-norm. If you wish to standardize, please use sklearn.preprocessing.StandardScaler before calling fit on an estimator with normalize=False

Copy_X: If True, X will be copied; else, it may be overwritten

n_jobs: The number of jobs to use for the computation. This will only provide speedup for n_targets > 1 and sufficient large problems. None means 1 unless in a joblib.parallel_backend context. -1 means using all processors
"""

#Fitting Linear Regression to dataset

from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

```python
def viz_linear():
    plt.scatter(X,y, color = 'blue')
    plt.plot(X, lin_reg.predict(X), color='red')
    plt.title("Linear Regression")
    plt.xlabel("$X$", fontsize = 18)
    plt.ylabel("$y$", fontsize = 18, rotation =0)
    plt.axis([0,2,0,15])
    plt.show()
    return
```

```python
viz_linear()
```
