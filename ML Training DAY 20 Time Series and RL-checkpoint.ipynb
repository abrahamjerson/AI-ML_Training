## Time Series Analysis

```python
#Sarima

import warnings
import itertools
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
data = sm.datasets.co2.load_pandas()
y = data.data #From object data, we are taking out the data to y
```

```python
#The 'MS' string groups the data in buckets by start of the month

y = y['co2'].resample('MS').mean()#Every month we take the mean

#The tern bfill means we use the value before filling missing values

y = y.fillna(y.bfill())

print(y)
```

```python
y.plot(figsize = (15,6))
plt.show()
```

```python
"""
Fitting an ARIMA Time Series Model
Using grid search we have identified the set of parameters that produces the best fitting model to our time series data
"""

mod = sm.tsa.statespace.SARIMAX(y, order =(1, 1, 1), seasonal_order =(1, 1, 1,12))
results = mod.fit()
```

```python
#Validating Forecasts

pred = results.get_prediction(start=pd.to_datetime('1998-01-01'))
```

```python
ax = y['1990':].plot(label='observed')
pred.predicted_mean.plot(ax = ax, label = 'One-step ahead Forecast', alpha=.7)
ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')
plt.legend()

plt.show()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['1998-01-01':]

#Compute the mean square error
mse = ((y_forecasted - y_truth) ** 2).mean()
print('The Mean Squared Error for our forecasts is {}'.format(round(mse,2)))
```

```python
#We specify to start computing the dynamic forecasts and confidence intervals form January 1998 onwards

pred_dynamic = results.get_prediction(start=pd.to_datetime('1998-01-01'), dynamic=True, full_results=True)
```

```python
"""
Plotting the observed and forecasted values of the time series, 
we can see that the overall forecasts are accurate even when using dynamic forcasts. 
All forecasted values (red line) match pretty close to the ground truth (blue line), and are well within the confidence intervals of our forecast.
"""
ax = y['1990':].plot(label='observed', figsize=(15,6))
pred_dynamic.predicted_mean.plot(label='Dynamic Forecast', ax=ax)

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show()
```

```python
#quantify the predictive perfomance of our forecasts by computing the MSE
#Extract the predicted and true values of our time series

y_forecasted = pred_dynamic.predicted_mean
y_truth = y['1998-01-01':]

#compute the mean square error
mse = ((y_forecasted - y_truth)**2).mean()
print('The Mean Squared Error for our forecasts is {}'.format(round(mse,2)))
```

```python
"""
Producting the Visualizing Forecasts
In the final step of this tutorial, we describe how to leverage our seasonal ARIMA time series model
to forecast future values. The get_forcast() attribute of our time series object can compute
forecasted values for a specified number of steps ahead
"""

#Get forcast 500 steps ahead in future

pred_uc = results.get_forecast(steps=500) #Here 500 steps means 500 month ahead prediction
pred_ci = pred_uc.conf_int()
```

```python
#The output of this code to plot the time series and forecasts to its future values

ax = y.plot(label = 'observed', figsize=(20, 15))

ax.fill_between(pred_ci.index,
                pred_ci.iloc[:,0],
                pred_ci.iloc[:,1], color='k', alpha=.2)

pred_uc.predicted_mean.plot(ax=ax, label='Forcast')

ax.set_xlabel('Date')
ax.set_ylabel('CO2 Levels')

plt.legend()
plt.show() #Shaded regions have the lowest and highest ranges for those values. As we move far ahead, the precision decreases
```

## Reinforcement Learning

```python
#Dependencies

from IPython.display import clear_output
import numpy as np
import random
import time
import gym
```

```python
#Environment

env = gym.make("Taxi-v3") #Create Environment
env.render() #To show the result. It is stochastic and not deterministic so changes everytime we run it
```

```python
#Q Table

#Number of possible actions
action_size = env.action_space.n
print("Action size:", action_size)

#Number of possible states
state_size = env.observation_space.n
print("State size:", state_size)
```

```python
qtable = np.zeros((state_size, action_size))
print(qtable)
```

```python
# Hyperparameters

episodes = 30000     # Total episodes
max_steps = 1000     # Max steps per episode
lr = 0.3             # Learning rate
decay_fac = 0.00001  # Decay learning rate each iteration
gamma = 0.90         # Discounting rate - later rewards impact less
```

```python
#Exploration Q learning

for episode in range(episodes):
    state = env.reset()    # Resets the environment
    done = False           # Are we done with the environment
    lr -= decay_fac         # Decay learning rate
    step = 0
    
    if lr<= 0: # Nothing more to learn
        break
    
    for step in range(max_steps):
        
        # Randomly chose an action
        action = env.action_space.sample()
        
        #Take the action -> Observe new state and reward
        new_state, reward, done, info = env.step(action)
        
        #Update qtable values
        if done == True: #If last, do not count future accumulated reward
            if (step < 1999 | step > 201):
                qtable[state, action] = qtable[state, action]+lr*(reward+gamma*0-qtable[state,action])
            break
            
        else: #Consider accumulated reward of the best decision stream
            qtable[state, action] = qtable[state, action]+lr*(reward+gamma*np.max(qtable[new_state,:])-qtable[state,action])
        
        #If done.. jump to next episode
        if done == True:
            break
        
        #Moving states
        state = new_state
        
    episode += 1
    
    if(episode % 3000 == 0):
        print('episode = ', episode)
        print('learning rate = ', lr)
        print('---------------')
        
```

```python
#Using Q-Table

#New Environment
state = env.reset()
env.render()
done = False
total_reward = 0

while(done == False):
    
    action = np.argmax(qtable[state,:]) #Chose best action (Q-table)
    state, reward, done, info = env.step(action) #Take action
    total_reward += reward #Summing reward
    
    #Display it
    time.sleep(0.5)
    clear_output(wait=True)
    env.render()
    print('Episodes Reward =', total_reward)
```
