```python
import scipy.stats as stats
import seaborn as sns
sns.set_context('notebook')
sns.set_style('darkgrid')
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
#Pandas Visualization

df = pd.DataFrame(np.random.randn(5,3), index=pd.date_range('1/1/2000', periods =5), columns =list('ABC')) # Random numpy vector arrays with 5 rows and 3 columns
df
```

```python

"""Scatter Plot
Scatter Plot can be created using the DataFrame.plot.scatter() method"""

df = pd.DataFrame (np.random.randn(50,3), columns = ['a', 'b', 'species']) #Here the random values are Normalized because or randn() method.
df #Mean = 0 and Variance = 1, So values from -1 to +1.
```

```python
df.plot.scatter(x='a', y='b', c='species', colormap='viridis')
plt.show()
```

```python
df = pd.DataFrame (np.random.rand(5,4), columns = ['a', 'b', 'c', 'd']) #Not Normalized because of rand() method
print(df)
print('\n')
```

```python
df.plot.bar()
plt.show()
```

```python
# To get horizontal plot use barh() method
df.plot.barh (figsize=(7,5))
plt.show()
```

```python
#To produce stacked bar plot, pass stacked = True
df.plot.bar(stacked=True)
plt.show()
```

```python
"""Histograms
Histograms can be plotted using plot.hist() method. We can specify number of bins"""
df = pd.DataFrame({'a':np.random.randn(1000)+1, 'b':np.random.randn(1000), 'c':np.random.randn(1000)-1}, columns = ['a', 'b', 'c',])
df.plot.hist(bins=10) #Bins are the number of bars in the plot.
plt.show()
```

```python
"""Area Plot
Area Plot can be craeted using the Series.plot.area() or the DataFrame.plot.area() methods"""

df = pd.DataFrame(np.random.rand(10,4), columns = ['a', 'b', 'c', 'd'])
df.plot.area()
plt.show()
```

```python
df = pd.DataFrame( 3* np.random.rand(4), index = ['a', 'b', 'c', 'd'], columns = ['x'])
df.plot.pie(subplots=True)
plt.show()
```
