```python
import numpy as np
from scipy import stats
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import poisson #poisson.rvs means taking a few random variables from poisson dist.
data_poisson = poisson.rvs(mu=10, size=2000) #mu is same as lamda.
```

```python
ax = sns.distplot(data_poisson, bins = 100, kde = True, color = 'skyblue', hist_kws = {'linewidth':1, 'alpha':1})
ax.set(xlabel = 'Poisson Distribution', ylabel = 'Frequency')
plt.plot
```
