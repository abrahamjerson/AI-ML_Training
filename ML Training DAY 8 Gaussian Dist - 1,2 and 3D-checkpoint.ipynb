```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.stats import multivariate_normal
from scipy.stats import norm
x = np.linspace(-5, 5, 50, endpoint=False)

#x is a univariate
#x takes values between -5 and 5 and there are 50 vales.
#Mean of y = 0 and variance = 0.5

y=norm.pdf(x, 0, 0.5);
print(y)
```

```python
plt.plot(x,y)
plt.show()
#When value of sigma square = 0, value goes upto infinity. Its is called dirac distribution.
```

```python
#Gaussian 2D

x,y = np.mgrid [-3:4:0.01, -2:2:0.01] #x=values from 3 to 4 with intervals 0.01 and same for y from -2 to 2
pos = np.empty(x.shape + (2,))
pos[:, :, 0] = x; pos[:, :, 1]=y
rv = multivariate_normal([0.5, -0.2], [[2, 0.5], [0.5,0.5]])#center at x=0.5, y=-0.2 and matrix of 2 dimension
plt.contourf(x, y, rv.pdf(pos))
plt.show()
```

```python
#Gaussian - 2D (Poly3D Collection)

from matplotlib import cm 
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()

ax = Axes3D(fig)

ax.plot_surface(X=x, Y=y, Z=rv.pdf(pos), rstride=4, cstride=4, cmap =cm.jet, linewidth =1, antialiased=True)
# x= above x,  y = above y and z = values from above pdf 
plt.show()
```
