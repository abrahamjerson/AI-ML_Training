```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import bernoulli
data_bern = bernoulli.rvs(size=10000, p=0.6)
```

```python
ax = sns.distplot(data_bern, kde=False, hist_kws={"linewidth":3, 'alpha':1})
ax.set(xlabel = 'Bernoulli Distribution', ylabel = 'Frequency')
plt.plot()
```

```python
ax = sns.distplot(data_bern, kde=False, hist_kws={"linewidth":3, 'alpha':1}, vertical = True)
ax.set(xlabel = 'Bernoulli Distribution', ylabel = 'Frequency')
plt.plot()
data_bern = bernoulli.rvs(size=10000, p=0.6)
```
