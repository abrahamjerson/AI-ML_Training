```python

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

#plt.rcParams.update({'figure.max_open_warning': 0})
```

```python
#First Dataset

titan = sns.load_dataset('titanic')
titan.head()
```

```python
fig, ax1 = plt.subplots (figsize=(7,5))
sns.barplot(x='who', y='survived', hue='class', data=titan, ax = ax1)
plt.show()
```

```python
fig, ax1 = plt.subplots (figsize=(7,5))
sns.barplot(y='who', x='survived', hue='class', data=titan, ax = ax1)
plt.show()
```

```python
plt.figure(figsize=(5,7))
sns.countplot ('deck', color = 'salmon', data = titan)
plt.show()
```

```python
#Second Dataset

tp = sns.load_dataset('tips')
tp.head() #Only displays a few lines of the Dataset
#tp - to dsiplay the full Dataset
```

```python
tp.describe() #Six point summarization
```

```python
#Plotting the Dataset
tp.plot()
plt.show()
```

```python
plt.subplots (figsize=(7,5))
sns.boxplot('day', 'total_bill', hue='smoker', data=tp)
plt.show()
```

```python
import warnings #It depends on the versions. Certain functions may be deprecated in certain versions.
warnings.filterwarnings ('ignore') #To remove the clutter on the screen or to ignore all the warnings by the compiler.

plt.figure(figsize =(4,3))
sns.factorplot(x='time', y='total_bill', hue ='smoker', col = 'day', data =tp, kind ='box', size = 5, aspect =0.5)
plt.show()
```

```python
plt.figure(figsize = (15,5))
#plt.figure()
sns.stripplot ('total_bill', 'day', data =tp, hue = 'time')
plt.show()
```

```python
plt.figure(figsize=(15,4))
#plt.figure()
sns.violinplot('total_bill', 'day', hue = 'time', data = tp, split = True)
plt.show()
```

```python
plt.figure(figsize=(15,4))
#plt.figure()
sns.violinplot('total_bill', 'day', hue = 'sex', data =tp, split = True)
sns.swarmplot('total_bill', 'day', hue='sex', color = 'w', data =tp, dodge =True)
plt.show()
```
