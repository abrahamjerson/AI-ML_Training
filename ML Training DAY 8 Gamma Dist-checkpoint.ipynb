```python
import numpy as np
import scipy.stats as stats

from matplotlib import pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import gamma
data_gamma = gamma.rvs (a=9, scale =1, size=10000)
```

```python
ax = sns.distplot(data_gamma, kde=True, bins = 100, color='maroon', hist_kws = {"linewidth": 15, "alpha": 1})
ax.set (xlabel = "Gamma Distribution", ylabel = "Probability")
plt.plot()
```
